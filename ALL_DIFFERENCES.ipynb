```python
dx = 0.05
x = linspace(0,10,100)

##Original function
t= x*exp(x)
#subplot(3,1,1)
#plot(x,t,'m-')

##exact derivative
#y_e = diff(t)
y_e = (x+1)*exp(x)
#subplot(3,1,2)
#plot(x[0:999],y_e[0:999],'b-.',x,y_f)
```

```python
y_f = (((x+dx)*exp(x+dx))-(x*exp(x)))/dx   #FORWARD DIFFERENCE
y_c = (((x+dx)*exp(x+dx))-((x-dx)*exp(x-dx)))/(2*dx) #CENTRAL DIFFERENCE
y_b = ((x*exp(x))-((x-dx)*exp(x-dx)))/dx      #BACKWARD DIFFERENCE
```

```python
subplot(2,2,1)
plot(x[0:99],t[0:99]/100000)
grid (True)
title('Agraph showing the original and differentiated functions')
xlabel('x')
ylabel('t')
```

```python
################################################################
subplot(2,2,2)
plot(x[0:99],y_e[0:99]/100000,'b-.', x[0:99],y_f[0:99]/100000,'g')
grid (True)
legend([['y_e'],['y_f']], loc ='best')
#title('Agraph showing the exact derivative and forward_difference functions')
#xlabel('X')
#ylabel('Y')
```

```python
###########################################################
subplot(2,2,3)
plot(x[0:99], y_e[0:99]/100000,'b-.',x[0:99],y_c[0:99]/100000,'m')
grid (True)
legend([['y_e'],['y_c']], loc ='best')
#title('Agraph showing the exact derivative and Central_difference functions')
#xlabel('X')
#ylabel('Y')

######################################################

subplot(2,2,4)
plot(x[0:99], y_e[0:99]/100000,'b-.',x[0:99],y_b[0:99]/100000,'r')
#legend((t,y_e,y_f),('Original','Exact derivative','Forward difference'))
grid (True)
legend([['y_e'],['y_b']], loc ='best')
#title('Agraph showing the exact derivative and backward_difference functions')
#xlabel('X')
#ylabel('Y')
```
