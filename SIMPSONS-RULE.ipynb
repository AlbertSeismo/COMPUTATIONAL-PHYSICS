```python
from numpy import *
from sympy import *
from sympy.abc import r

#Lets define the Simpson Rule constants in the formula
A =0      #initial coordinate
B=3       #Final coordinate
N=397       #Number of rectangles (N must be ODD), note that the higher the value of N, the more accurate the final answer
h = (B-A)/(N-1) #width of each rectangle
```

```python
#####Lets start writing a function that includes that calls the function to be integrated

def f(x):
    f = x*x
    print('x f(x) =',x,x*x)
    return (f)
##########################################################No need to change this part of the code-start
def Simps_ih(i,h):
    if ((i ==1) or (i==N)):
        width =h/3
        return (width)
    elif (i%2==0):
        width = (4*h)/3
        return (width)
    else:
        width = (2*h)/3
        return (width)
```

```python
##Initialising the Area
SUM_A = 0.0

###Loop for calculating the x coordinates(x0,x1,x2,x3,x4)
for i in range(1,N+1):
    t = A+ (i-1)*h
    #print(t)
    #print(f(t))
    w = Simps_ih(i,h)
    #print(w)
    SUM_A = SUM_A + w*f(t)  #This line calculates the total sum of the area by Trapezoidal Rule
print('\nTotal Area =',SUM_A)

##########################################################No need to change this part of the code-end

#####################Finding exact integral and compare
```

```python

F = Integral(r**2,(r,0,3))
F.as_sum(4,method ='midpoint')
A_ex = F.n()
print('\nThe exact area by integration of a function is =',A_ex)

DeltaA = abs(A_ex-SUM_A)
print('\nThe error made by integrating a fuction using... \nSimpson rule of {} rectangles and exact solution is = {}\n'.format(N,DeltaA))
```
